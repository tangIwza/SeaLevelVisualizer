```python
import pdfplumber

pdf_file = r"C:\Users\2026trainees69\Downloads\MT2026.pdf"

# Create a list to hold 12 months, where each month is a list of daily data
monthly_tide_data = []

with pdfplumber.open(pdf_file) as pdf:
    for page in pdf.pages:
        text = page.extract_text()
        
        # Check if the text indicates this page is for Map Ta Phut
        if text and "Map Ta Phut" in text:
            month_data = []
            for line in text.splitlines():
                parts = line.strip().split()
                
                # Check if the line starts with a day number
                if parts and parts[0].isdigit():
                    # Filter out non-numeric characters (like moon phase symbols)
                    numeric_parts = []
                    for p in parts:
                        try:
                            float(p) # This works for integers and decimals like '1.2' or '-0.1'
                            numeric_parts.append(p)
                        except ValueError:
                            pass
                    
                    # We need 1 day + 24 hours = 25 values
                    if len(numeric_parts) >= 25:
                        month_data.append(numeric_parts[:25])
            
            # If we successfully extracted days for this page, add it to our months list
            if month_data:
                monthly_tide_data.append(month_data)

# Print a summary to verify we have 12 months
#print(f"Total months extracted: {len(monthly_tide_data)}")
#for i, month in enumerate(monthly_tide_data):
 #   print(f"Month {i+1}: {len(month)} days")

# Show sample data for the first 3 days of Month 1

for j in range(0,12):
    print("\nSample Data (Month ",j+1,", first 3 days):")
    for day_data in monthly_tide_data[j][:3]:
        print(day_data)
```

```python
import json
output_file = r"C:\Users\2026trainees69\Desktop\SeaLevelVisualizer\public\tide_data.json"

json_data = []
for m_idx, month in enumerate(monthly_tide_data):
    month_obj = {
        "month": m_idx + 1,
        "days": []
    }
    for day_data in month:
        month_obj["days"].append({
            "day": int(day_data[0]),
            "hours": [float(h) for h in day_data[1:25]]
        })
    json_data.append(month_obj)

with open(output_file, 'w', encoding='utf-8') as f:
    json.dump(json_data, f, indent=2)

print(f"Successfully saved full data to {output_file}")
```
